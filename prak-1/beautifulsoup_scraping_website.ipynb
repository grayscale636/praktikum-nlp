## Single Link

```python
import requests
import html5lib
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link = input("Masukkan link artikel: ")
req = requests.get(link)
soup = bs(req.content, 'html5lib')
```

```python
paragraphs = soup.findAll('p')
```

```python
text = []
for p in paragraphs:
  text.append(p.text)
```

```python
text_ready = ' '.join(text)
text_ready
```

```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters and digits
    text = re.sub(r"\s+[^a-zA-Z]\s", " ", text)

    # Remove extra spaces
    text = re.sub(' +', ' ', text)

    # Remove leading and trailing spaces
    text = text.strip()

    return text

text = text_ready
text_siap = preprocess_text(text)
```
